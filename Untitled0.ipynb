```python
conversion_rates = {
    'usd': 1,
    'cad': 0.78,
    'dkk': 0.16,
    'lpa': 0.013,
    'aud': 0.70,
    'eur': 1.10,
    'gbp': 1.38,
    'sgd': 0.74,
    'chf': 1.08,
    'inr': 0.013,
}

def extract_numeric_value(salary):
    # Extract the numeric value using regex
    match = re.search(r'[\d,.]+', str(salary))
    return match.group() if match else np.nan

def clean_salary(salary):
    if pd.isna(salary):
        return np.nan

    # Convert to lower case and ensure it's a string
    salary = str(salary).lower()

    # Extract currency if present
    currency = 'usd'  # Default currency
    for cur in conversion_rates.keys():
        if cur in salary:
            currency = cur # Fixed typo: cursalary -> currency
            salary = salary.replace(cur, '')
            break

    # Handle 'k' (thousands)
    if 'k' in salary:
        salary = salary.replace('k', '')
        try:
            salary = float(salary) * 1000
        except ValueError:
            return np.nan

    # Extract numeric value from the salary string
    salary = extract_numeric_value(salary)

    # Handle commas and convert to float
    salary = str(salary).replace(',', '')
    try:
        salary = float(salary)
    except ValueError:
        salary = np.nan

    # Convert to USD
    if currency in conversion_rates:
        salary = salary * conversion_rates[currency]

    return salary

# Apply the clean_salary function to the 'salary' column and create a new column 'cleaned_salary' - moved outside the function
df['cleaned_salary'] = df['salary'].apply(clean_salary)

# Display the cleaned DataFrame
print(df)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['cleaned_salary'].dropna(), bins=30, edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
